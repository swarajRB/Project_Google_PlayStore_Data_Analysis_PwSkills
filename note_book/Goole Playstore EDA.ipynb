```python
df.Size.unique()
```

```python
# Writing a simple code to filter and convert the size

# 1MB = 1024kb

def size_process(item):
    if str(item)[-1] == "M":
        res = float(str(item).replace("M",""))
        res = res*1024
        return res
    elif str(item)[-1] == "k":
        res = (str(item).replace("k",""))
        return res
    else:
        return str(np.nan)
```

```python
df_copy["Size"] = df_copy["Size"].apply(size_process)
```

```python
df_copy["Size"].unique()
```

```python
df_copy['Size'].dtypes
```

```python
# Converting to float from object type
df_copy['Size'] = df_copy.Size.astype('float')

df_copy['Size'].dtypes
```

```python
# Now to work on "Installs" columns
df_copy['Installs']
```

```python

df_copy['Installs'] = df_copy['Installs'].str.replace('+','').str.replace(',','')
df_copy['Installs'] = df_copy['Installs'].astype(int)
```

```python
df_copy.info()
```

```python
# Now Cleaing the "Price" columns
df_copy['Price'] = df['Price'].str.replace('$','')
```

```python
# Converint the Type from String to float and int
df_copy['Price'] = df_copy['Price'].astype(float)
```

```python
df_copy['Price'].value_counts()
```

```python
df_copy.info()
```

```python
# There is column as last updated as in the form of object
# We can covert it into date_time format and separate columns of day,month and year

df_copy['Last Updated']
```

```python
df_copy['Last Updated'] = pd.to_datetime(df_copy['Last Updated']) # Converted to date time format

df_copy['Last Updated']
```

```python
# Making separat columns for Day, Month and Year

df_copy['Day'] = df_copy['Last Updated'].dt.day # For Day Columns
df_copy['Month'] = df_copy['Last Updated'].dt.month # For Month column
df_copy['Year']  = df_copy['Last Updated'].dt.year # For Year Column
```

```python
# Drop rows with missing values in critical columns
df_copy.dropna(subset=['Type'], inplace=True)

# Fill missing values in numerical columns
df_copy['Rating'].fillna(df_copy['Rating'].median(), inplace=True)
df_copy['Reviews'].fillna(df_copy['Reviews'].median(), inplace=True)

# Fill missing values in categorical columns
df_copy['Current Ver'].fillna(df_copy['Current Ver'].mode()[0], inplace=True)
df_copy['Android Ver'].fillna(df_copy['Android Ver'].mode()[0], inplace=True)

df_copy['Size'].fillna(df_copy['Size'].median(), inplace=True)
df_copy['Price'].fillna(0, inplace=True)
```

```python
# Final check
print("Missing values after imputation:")
print(df_copy.isnull().sum())
```

## 📊 Data Visualization Techniques

When analyzing data, different types of visualizations are used for **categorical** and **numerical** data.


### 🔹 **For Categorical Data:**

These charts help **understand frequency distribution** and comparisons across categories.
- **Frequency Plot** – Shows how often each category appears.
- **Bar Chart** – Compares different categories using bars.
- **Pie Chart** – Represents proportions of categories in a circular chart.
- **Column Chart** – Similar to a bar chart but with vertical bars.
- **Clustered Chart** – Groups related categories together for better comparison.


### 🔹 **For Numerical Data:**

These charts help **understand distributions, trends, and relationships**.
- **Histogram** – Displays the distribution of numerical values.
- **Distplot** – A density plot that combines a histogram and a smooth curve.
- **Box Plot** – Shows median, quartiles, and outliers in numerical data.
- **Line Chart** – Helps track trends over time.
- **Pair Plot** – Displays relationships between multiple numerical variables.

```python
# Function to plot numerical variable distributions
def plot_numerical_univariate(column):
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))

    # Histogram
    sns.histplot(df_copy[column].dropna(), bins=20, kde=True, ax=axes[0], color="blue")
    axes[0].set_title(f"Histogram of {column}")

    # Box Plot
    sns.boxplot(y=df_copy[column], ax=axes[1], color="orange")
    axes[1].set_title(f"Box Plot of {column}")

    # Distplot
    sns.kdeplot(df_copy[column].dropna(), fill=True, ax=axes[2], color="green")
    axes[2].set_title(f"Distribution Plot of {column}")

    plt.show()

# Example: Analyzing 'Rating' column
plot_numerical_univariate('Rating')
```
